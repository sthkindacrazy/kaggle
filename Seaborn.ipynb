```python
%matplotlib inline
#visualization result를 바로 띄우기 위해서 (default는 file로 저장) jupyter 전용 notation

import pandas as pd
import seaborn as sns
```

## Load Dataset

```python
train = pd.read_csv("/Users/Joohyung/Desktop/kaggle/titanic/train.csv", index_col="PassengerId")
print(train.shape)
train.head()
```

## Seaborn

### countplot 

```python
sns.countplot(data=train, x="Embarked", hue="Survived")
```

```python
sns.countplot(data=train, x="Pclass", hue="Survived")
```

### Barplot

```python
#sns.barplot(data=train, x="Pclass", y="Fare")
#운임요금의 평균값 (y축으로 나타낼 수 있음 countplot -> row만 세고 싶다. column의 평균을 알고 싶다 barplot)
sns.barplot(data=train, x="Pclass", y="Fare", hue="Survived")
```

### distplot (분포)

```python
sns.distplot(train["Fare"], hist=False)
```

```python
ignore_high_fare = train[train["Fare"] < 100]

sns.distplot(ignore_high_fare["Fare"], hist=False)
```

```python
sns.countplot(data=train, x="SibSp", hue="Survived")
```

```python
perish = ignore_high_fare[ignore_high_fare["Survived"] == 0]
survived = ignore_high_fare[ignore_high_fare["Survived"] == 1]

sns.distplot(perish["Fare"], hist=False, label="Perish")
sns.distplot(survived["Fare"], hist=False, label="Survived")
```

## lmplot 

```python
sns.lmplot(data=train, x="Age", y="Fare", fit_reg=False, hue="Survived")
```

```python
ignore_high_fare = train[train["Fare"] < 100]
sns.lmplot(data=ignore_high_fare, x="Age", y="Fare", fit_reg=False, hue="Survived")
```
